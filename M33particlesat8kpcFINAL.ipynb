```python
def OrbitCOM(galaxy,start,end,n):
    """function that loops over all the desired snapshots to compute the COM pos and vel as a function of time.
    inputs: galaxy - the name of the galaxy
            start - the first snapshot to read into the program
            end - the last snapshot to be read in
            n - an interval for the loop
          
    returns: time
             Average Position of the solar analogues [kpc]
             velocity of the center of mass [km/s]
             """

    
    # compose the filename for output
    
    fileout = 'AnalogueOrbit_'+'%s'%(galaxy)+'.txt'
    
    #we have to define the properties of the galaxy for our calculations
    delta = 0.1
    VolDec = 4 
    
    # generate the snapshot id sequence 
    snap_ids = np.arange(start,end+1,n) #I added +1 because otherwise it wouldnt go to 800, it would stop at 795
    
    # initialize the array for the info we are exporting to graph
    orbit = np.zeros([len(snap_ids),5]) 
    # initialize the array for the position since we have to find the average
    r = np.zeros(200)

    for  i,snap_id in enumerate(snap_ids):# loop over files
        
        # compose the data filename (be careful about the folder)
        # add a string of the filenumber to the value "000"
        ilbl = '000' + str(snap_id)
        ilbl = ilbl[-3:]
        filename="%s_"%(galaxy) + ilbl + ".txt"
        
        #now we need to get the center of mass so that we can pick particles 8kpc away from the center of mass
        M33COM = CenterOfMass(filename,2)
        pos = M33COM.COM_P(0.1,4)
        vel = M33COM.COM_V(pos[0],pos[1],pos[2]).value
        pos = pos.value 
        
        #now we can define the particles that we are looking at for this simulation
        M33anx = M33COM.x[index] - pos[0]
        M33any = M33COM.y[index] - pos[1]
        M33anz = M33COM.z[index] - pos[2]

        #now we calculate the average distance from the center of mass
        r = np.sqrt(M33anx**2+M33any**2+M33anz**2)
        ravg = np.sum(r)/71
    
        # store the time, pos, vel in ith element of the orbit array,  without units (.value) 
        orbit[i] = M33COM.time.value/1000,ravg,*tuple(vel)
        
        # print snap_id to see the progress
        print(snap_id)
        
    # write the data to a file
    np.savetxt(fileout, orbit, fmt = "%11.3f"*5, comments='#',
               header="{:>10s}{:>11s}{:>11s}{:>11s}{:>11s}"\
                      .format('t', 'r', 'vx', 'vy', 'vz'))
    return orbit
```

Now I am adding a code to determine the average mass of the galaxy per year. I will calculate this by determining the amount of stars that become unbound. I will then find the ratio of bound stars to unbound stars to determine the mass loss rate of the galaxy

```python
def HernquistPotential(radius,a,mass):
    #function to calculate the Hernquist potential function
    #inputs:
        #r - distance from the center of the galaxy [kpc]
        #a - the scale radius [kpc]
        #mass - the mass of the dark matter halo in Msun
    #outputs:
        #value of the Hernquist profile
    #the escape velocity of a galaxy is vesc = sqrt(2*HP)
    G = const.G.value
    return -G*mass*1e12/(radius+a)
```
